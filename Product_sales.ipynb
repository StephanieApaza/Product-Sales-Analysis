Before cleaning and validation, the dataset contained 15000 rows and 8 columns. I validated all columns against the criteria in the dataset description.

* week: Six values (1 to 6), consistent with the description. No cleaning was needed.
* sales_method: Cleaned to standardize naming (e.g., email, Email + Call, em + call). After cleaning, only three sales methods remain, as described.
* customer_id: Unique values without missing records, consistent with the description. No cleaning was needed.
* nb_sold: Numeric values, consistent with the description. No cleaning was needed.
* revenue: Identified 1074 null values (7.16% of total). Since removing these records would have exceeded the 5% threshold (750 rows), which could reduce the sample and bias the analysis, missing values were imputed using the median revenue within each sales_method group. This approach ensures consistency with the typical behavior of each sales strategy while reducing the impact of outliers.
* years_as_customer: Cleaned, since the company was founded in 1984 and the current year is 2025, meaning the maximum possible value is 41 years. Out of range values were corrected, making the column consistent with the description.
* nb_site_visits: Numeric values without missing records, consistent with the description. No cleaning was needed.
* state: 50 unique customer locations without missing records, consistent with the description. No cleaning was needed.

After validation, the dataset contains 15000 rows and 8 columns with no missing values.

# **Business Metrics**

```python
avg_revenue_method = sales_df.groupby('sales_method')['revenue'].mean()
avg_revenue_method
```

```python
avg_revenue_call = avg_revenue_method.loc['Call']
avg_revenue_email = avg_revenue_method.loc['Email']
avg_revenue_email_call = avg_revenue_method.loc['Email and Call']
```

```python
revenue_per_minute_call = avg_revenue_call / 30
revenue_per_minute_email = avg_revenue_email / 5
revenue_per_minute_email_call = avg_revenue_email_call / 10
```

```python
revenue_per_minute_call.round(2)
```

```python
revenue_per_minute_email_call.round(2)
```

```python
revenue_per_minute_email.round(2)
```

Since our goal is to identify the most effective method to sell the new product, I recommend using **revenue per minute of effort** as our primary business metric. This metric allows us to compare sales methods not only in terms of average revenue generated but also by the time invested by the sales team. By prioritizing the method with the highest revenue per minute, we ensure an optimal balance between sales growth and efficient resource allocation.

Based on the current data, the results are as follows:
* Call: 1.59 revenue/minute
* Email and Call: 18.38 revenue/minute
* Email: 19.40 revenue/minute*

For Email, I assumed 5 minutes per email preparation and delivery. A more precise conclusion will be possible once additional data on actual preparation time is available.

# **Recommendation:**

Based on the analysis, I recommend focusing on the following steps:

1. **Prioritize Email and Call as the primary sales method**
This method shows a strong positive impact on revenue, a clear growth trend over time, and balances efficiency. Each case only requires sending an email plus a 10 minute follow-up call, which makes it both effective and manageable for the sales team.

2. **Monitor efficiency using revenue per minute of effort**
This metric highlights that Email and Call delivers high returns relative to the time invested. In contrast, the Call method requires 30 minutes per customer and generates the lowest revenue, making it the least efficient option.

3. **Leverage the Email method as a complementary strategy**
While slightly less efficient than Email and Call, Email alone reaches the largest number of customers, generates the most site visits, and leads to the highest number of products sold. Combining both methods (Email and Call + Email) could maximize overall reach and revenue.

4. **Improve data collection and quality**
Address missing values in revenue and gather more consistent data to enable deeper analysis and more accurate decision-making in the future.
